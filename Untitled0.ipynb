```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate Lissajous figures
def lissajous_figure(A, B, omega1, omega2, delta, t):
    x = A * np.sin(omega1 * t + delta)
    y = B * np.sin(omega2 * t)
    return x, y

# Parameters for Lissajous figures
A = 1
B = 1
omega1 = 1
omega2 = 1
t = np.linspace(0, 2 * np.pi, 1000)

# Lissajous figures of rank 1 (omega1/omega2 = 1)
plt.figure(figsize=(12, 6))
phases = [0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi]
for i, delta in enumerate(phases):
    x, y = lissajous_figure(A, B, omega1, omega2, delta, t)
    plt.subplot(1, len(phases), i + 1)
    plt.plot(x, y)
    plt.title(f'Delta = {delta:.2f} pi')
    plt.xlabel('x')
    plt.ylabel('y')
plt.suptitle('Lissajous Figures (Rank 1)')
plt.tight_layout()
plt.show()

# Lissajous figures of rank 2 (omega1/omega2 = 2)
omega1_rank2 = 2
omega2_rank2 = 1
plt.figure(figsize=(12, 6))
for i, delta in enumerate(phases):
    x, y = lissajous_figure(A, B, omega1_rank2, omega2_rank2, delta, t)
    plt.subplot(1, len(phases), i + 1)
    plt.plot(x, y)
    plt.title(f'Delta = {delta:.2f} pi')
    plt.xlabel('x')
    plt.ylabel('y')
plt.suptitle('Lissajous Figures (Rank 2)')
plt.tight_layout()
plt.show()


# Chebyshev polynomial (order n=5)
n = 5
x_cheb = np.linspace(-1, 1, 1000)
y_cheb = np.cos(n * np.arccos(x_cheb))

plt.figure()
plt.plot(x_cheb, y_cheb)
plt.title(f'Chebyshev Polynomial of Order {n}')
plt.xlabel('x')
plt.ylabel('T_n(x)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the characteristic polynomial
def D(p):
    return p**77 + p**55 + p**4 + p + 3

# Generate values of p along the imaginary axis (p = j*omega)
omega = np.linspace(-10, 10, 1000) # Adjust the range as needed
p = 1j * omega

# Evaluate the polynomial D(p) for these values
D_values = D(p)

# Plot the Mikhailov hodograph (real part vs. imaginary part of D(p))
plt.figure(figsize=(8, 6))
plt.plot(D_values.real, D_values.imag)
plt.xlabel('Real part of D(j*omega)')
plt.ylabel('Imaginary part of D(j*omega)')
plt.title('Mikhailov Hodograph')
plt.grid(True)
plt.axhline(0, color='k', linestyle='-', linewidth=0.5)
plt.axvline(0, color='k', linestyle='-', linewidth=0.5)
plt.show()

# Analyze stability using the Mikhailov criterion
# The system is stable if the hodograph starts on the positive real axis
# and encircles the origin N/2 times in the counter-clockwise direction,
# where N is the degree of the polynomial (N=77 in this case).
# We need to examine the plot to determine stability based on this criterion.
print("Examine the Mikhailov hodograph plot to determine the stability of the system.")
print("For stability, the hodograph should start on the positive real axis and encircle the origin N/2 times counter-clockwise.")
print(f"The degree of the polynomial is N = 77.")
```

```python
import random

# Initialize ATM with supported card types and codes
supported_cards = {
    1: [1111, 2222],  # Card Type 1 with supported codes
    2: [3333, 4444]   # Card Type 2 with supported codes
}

# Initialize account balance with a random value (integer hryvnia and kopecks)
account_balance = round(random.uniform(100.00, 10000.00), 2)
print(f"Account balance initialized to: {account_balance:.2f} UAH")

# ATM simulation states
def state_a0():
    """a0: Запрошення до вводу картки."""
    print("\nБудь ласка, вставте картку.")
    # Simulate card insertion and transition to a1
    return state_a1

def state_a1():
    """a1: Розпізнання картки."""
    # Simulate card recognition. In a real ATM, this would involve reading card data.
    # For this simulation, we'll ask the user to input card type and code.
    try:
        card_type = int(input("Введіть тип картки (1 або 2): "))
        card_code = int(input("Введіть код картки: "))

        if card_type in supported_cards and card_code in supported_cards[card_type]:
            print("Картка розпізнана.")
            return state_a2(card_type, card_code) # Pass card info to the next state
        elif card_type not in supported_cards:
            print("Непідтримуваний тип картки.")
            return state_a13
        else:
            print("Невірний код картки.")
            return state_a11

    except ValueError:
        print("Невірний ввід. Спробуйте знову.")
        return state_a1


def state_a2(card_type, card_code):
    """a2: Запрошення до вводу коду."""
    print("Будь ласка, введіть ваш PIN-код.")
    return state_a3(card_type, card_code)

def state_a3(card_type, card_code, attempts=3):
    """a3: Перевірка коду."""
    pin = input("Введіть PIN-код: ")
    # In a real ATM, this would compare the entered PIN with the one on the card/server.
    # In this simulation, we compare with the entered card_code for simplicity.
    try:
        if int(pin) == card_code:
            print("PIN-код прийнято.")
            return state_a4
        else:
            attempts -= 1
            if attempts > 0:
                print(f"Невірний PIN-код. Залишилось спроб: {attempts}")
                return state_a3(card_type, card_code, attempts)
            else:
                print("Введено невірний PIN-код тричі. Картка заблокована.")
                return state_a12
    except ValueError:
        attempts -= 1
        if attempts > 0:
            print(f"Невірний ввід. Залишилось спроб: {attempts}")
            return state_a3(card_type, card_code, attempts)
        else:
            print("Невірний ввід тричі. Картка заблокована.")
            return state_a12


def state_a4():
    """a4: Вивід меню."""
    print("\nОберіть операцію:")
    print("1. Стан рахунку")
    print("2. Зняти гроші")
    print("3. Вихід")
    choice = input("Введіть номер операції: ")

    if choice == '1':
        return state_a5
    elif choice == '2':
        return state_a6
    elif choice == '3':
        return state_a10
    else:
        print("Невірний вибір. Спробуйте знову.")
        return state_a4

def state_a5():
    """a5: Вивід стану рахунку."""
    global account_balance
    print(f"\nВаш поточний баланс: {account_balance:.2f} UAH")
    return state_a4

def state_a6():
    """a6: Запит на введення суми."""
    try:
        amount = float(input("Введіть суму для зняття: "))
        if amount <= 0:
            print("Сума для зняття повинна бути більше нуля.")
            return state_a6
        return state_a7(amount)
    except ValueError:
        print("Невірний ввід суми. Спробуйте знову.")
        return state_a6


def state_a7(amount):
    """a7: Перевірка наявності грошей."""
    global account_balance
    if account_balance >= amount:
        return state_a8(amount)
    else:
        return state_a9

def state_a8(amount):
    """a8: Видача готівки та корегування рахунку."""
    global account_balance
    account_balance -= amount
    print(f"\nВидано готівку: {amount:.2f} UAH")
    print(f"Залишок на рахунку: {account_balance:.2f} UAH")
    return state_a4

def state_a9():
    """a9: Повідомлення про недостатню кількість грошей."""
    print("\nНедостатньо коштів на рахунку.")
    return state_a4

def state_a10():
    """a10: Прощальні вітання."""
    print("\nДякуємо за користування банкоматом. До побачення!")
    return None # Exit the simulation

def state_a11():
    """a11: Запрошення до повторного вводу коду."""
    print("\nБудь ласка, спробуйте ввести код ще раз.")
    # In this simulation, we'll just go back to state a1 for simplicity
    # A real ATM would handle this differently (e.g., track attempts per card)
    return state_a1

def state_a12():
    """a12: Повідомлення про невірний код і блокування картки."""
    print("\nВаша картка заблокована. Зверніться до банку.")
    return state_a0 # Go back to the initial state for the next user

def state_a13():
    """a13: Повідомлення, що запропонований тип картки не підтримується."""
    print("\nВибачте, даний тип картки не підтримується.")
    return state_a0 # Go back to the initial state for the next user


# Main ATM simulation loop
current_state = state_a0

while current_state is not None:
    current_state = current_state()
```
